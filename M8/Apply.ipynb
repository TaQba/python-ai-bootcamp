```python
import pandas as pd
df = pd.read_excel('Pivot.xlsx')
df.head()
```

```python
def commission_fee(x):
    if x <= 300:
        return 0
    elif x <= 900:
        return x * 0.03
    else:
        return x * 0.06
df['commission_fee'] = df['Sprzedaż'].apply(lambda x: commission_fee(x))
df['Produkt_len'] = df['Produkt'].apply(len)
```

```python
df.head()
```

```python
import math
df['#_opakowań'] = df['Sztuki'].apply(lambda x: math.ceil(x/5))
df.head()
```

```python
#Tym razem funkcja, którą będziemy podawać dalej metodzie apply, jako argument będzie przyjmowała cały wiersz wartości

def bonus(row):
    margin = (row['Sprzedaż'] - row['Koszty'])/row['Sprzedaż']
    if margin > 0.55:
        return 200
    else:
        return 0
df['Bonus'] = df.apply(lambda row: bonus(row),axis=1)
df.sample(10)
```

Map

```python
car_dict = dict(zip(df['Przedstawiciel'].unique(),['Mazda','Toyota','BMW','Audi','Fiat','Seat']))
```

```python
#Użyjemy teraz funkcji map i stworzonego słownika, aby dodać kolumnę przechowującą informację o marce samochodu prowadzonego przez każdego z przedstawicieli.
df['Marka_samochodu'] = df['Przedstawiciel'].map(car_dict)
```

```python
df.head()
```

Applymap

```python
#Funkcji applymap używać będziemy w sytuacji wymagającej wykonania tej samej transformacji na wielu kolumnach i wierszach jednocześnie. Możemy jej używać wyłącznie na obiektach DataFrame, a jedynym akceptowalnym argumentem jest funkcja, którą możemy wywołać na każdym elemencie z osobna.
df.applymap(lambda x: x.upper() if isinstance(x,str) else x)
df.head()
```
