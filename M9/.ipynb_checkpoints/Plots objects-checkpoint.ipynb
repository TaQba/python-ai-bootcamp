# Plots - Objects way

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

fig = plt.figure()
axes = fig.add_axes([0,0,1,1]) # left, bottom, width, height
x = np.arange(11)
y = x ** 2
axes.plot(x,y)
axes.set_xlabel('oś X')
axes.set_ylabel('oś Y')
axes.set_title('Tytuł wykresu')
```
