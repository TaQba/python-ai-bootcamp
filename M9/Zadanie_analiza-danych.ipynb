### 2. Jakie źródła pozyskania pracownika (Recruitment Source) są najlepsze, jeśli zależy nam na jak najdłuższym stażu pracowników? 

```python
df[['DateofTermination','DOB','DateofHire']].head()
# df['DOB'] = pd.to_datetime(df['DOB'],format='%m/%d/%Y')
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'],format='%m/%d/%y')
df['DateofHire'] = pd.to_datetime(df['DateofHire'],format='%m/%d/%Y')
def fix_year(date_str):
    dt = pd.to_datetime(date_str, format='%m/%d/%y')
    if dt.year > 2025:  # or any cutoff logic
        dt = dt.replace(year=dt.year - 100)
    return dt
df['DOB'] = df['DOB'].apply(fix_year)
```

```python
import datetime as dt

def count_seniority(row):
    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2019, 9, 27)
    else:
        end_date = row['DateofTermination']
    
    delta = end_date - row['DateofHire']
    return delta.days / 365.25 

df['Seniority'] = df.apply(lambda row: count_seniority(row),axis=1)

df.head()
```

```python
plt.figure(figsize=(55,5))
sns.barplot(x='RecruitmentSource',y='Seniority',data=df)
```

#### CarrieBuilder najlepszy !

###  5. Czy starsi pracownicy pracują nad większą liczbą specjalnych projektów niż młodsi pracownicy?

```python
def is_senior(row):

    dt = pd.to_datetime(row['DOB'], format='%y/%m/%d')
    if dt.year > 1980:
        return 0
    return 1
current['Senior'] = df.apply(lambda row: is_senior(row),axis=1)
```

```python
seniors = current.groupby(['Senior']).count().reset_index().copy()
seniors
```

```python
sns.barplot(x='Senior',y='SpecialProjectsCount',data=seniors)
```

#### Starci (czyli urodzenii do 1980) pracuja nas mniejszą iloscia specjalnych projektow nie młodsi
