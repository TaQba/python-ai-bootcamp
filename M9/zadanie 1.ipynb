# Zadanie: baza filmów

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_movies = pd.read_csv('tmdb_movies.csv')
df_genres = pd.read_csv('tmdb_genres.csv')
```

```python
df_movies.sort_values(by='vote_count',ascending=False).head(2)
```

### 1. Zwróć listę 10 najwyżej ocenianych filmów (vote_average), których liczba głosów (vote_count) jest większa od 3. kwartyla rozkładu liczby głosów.

```python
q3 = len(df_movies.index)*0.75 #3rd quartile
df_movies[df_movies['vote_count'] > q3].sort_values(by='vote_average',ascending=False).head(10)
```

#### 2.Pogrupuj tabelę w taki sposób, aby otrzymać średni przychód (revenue) oraz średni budżet (budget) w danym roku dla filmów opublikowanych od 2010 (włącznie) do 2016 roku (włącznie). Następnie na tej podstawie stwórz wykres, w którym średnie przychody są wykresem kolumnowym, a średni budżet wykresem liniowym na tych samych osiach. Sformatuj odpowiednio oś X oraz oś Y. Dodaj tytuł wykresu, oraz legendę, która znajduje się w prawym górnym rogu płótna, lecz poza obszarem osi. Przykład wykresu widoczny poniżej:

```python
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'], errors='coerce') # fix data format
df_movies['release_year'] = df_movies['release_date'].dt.year #add col release_year

df_movies =  df_movies[(df_movies['release_year'] >= 2010) & (df_movies['release_year'] <= 2016)]

df_movies_budzet = df_movies.groupby('release_year').agg({'budget':'mean'}).reset_index()
df_movies_revenue = df_movies.groupby('release_year').agg({'revenue':'mean'}).reset_index()


def million(x, pos):
        return '{}M'.format(x*1e-6)
    
fig, ax = plt.subplots()
formatter = plt.FuncFormatter(million)
ax.yaxis.set_major_formatter(formatter)
ax.bar(df_movies_revenue['release_year'], df_movies_revenue['revenue'], label='revenue')
ax.plot(df_movies_budzet['release_year'], df_movies_budzet['budget'], color='red', label='budget')
plt.title("Średni przychód i budżet filmów w latach 2010-2016 roku")
ax.legend(loc=(1.05,0.8))
```

#### 3. Baza filmów zawiera kolumnę z id gatunku (genre_id). Na tej podstawie połącz ze sobą bazę filmów z bazą gatunków, tak aby w bazie filmów można było odczytać nazwę gatunku filmu.


```python
df_movies.head()
df_genres = df_genres.rename(columns={'Unnamed: 0': 'genre_id'})
#polacznie baz
merged_df = pd.merge(df_movies, df_genres, on='genre_id', how='inner') 
merged_df['per_genre'] = 1
merged_df.head()
```

#### 4. Jaki gatunek filmu z bazy pojawia się w niej najczęściej? Ile filmów tego gatunku znajduje się w bazie?

```python
top_genre = merged_df.groupby('genres').agg({'per_genre':'count'}).reset_index().head(1)
top_genre
```

#### 5. Filmy, którego gatunku trwają średnio najdłużej (runtime)?

```python
merged_df.groupby('genres').agg({'runtime':'mean'}).reset_index().head(1)
```

#### 6. Stwórz histogram czasu trwania filmów z gatunku, który cechuje się największym średnim czasem trwania.

```python
merged_df[merged_df['genres'] == top_genre['genres'][0]]['runtime'].hist(bins=30)
```
