```python
y=5 #liczba lat
m=12 #liczba miesiecy
flat_value = 120000 #cena mieszkania
flat_ppy = 0.05 #wzrost roczny procentowy ceny mieszkani
base_rate = 0.12
```

```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

```python
vector_years = np.arange(1,y+1)
```

```python
vector_months = np.arange(1,(y*m)+1) # wektor z liczba miesiecy w 5 latach
```

```python
vector_flat_price_years =  flat_value * ((1+flat_ppy)**vector_years) 
```

```python
print("Pytanie 1: Cena mieszkania po 5 latach: {:0.2f}zł".format(vector_flat_price_years[-1]))
```

```python
monthly_rate = - npf.pmt((base_rate/m), m*y, flat_value)
```

```python
print("Pytanie 2: Miesieczna rata wynosi: {:0.2f}zł".format(monthly_rate))
```

```python
vector_flat_price_months_in_k = flat_value * ((1+flat_ppy/12)**vector_months)/1000
rounded = np.rint(vector_flat_price_months_in_k).astype(int)
```

```python
vector_savings_k = (vector_months * monthly_rate)/1000
vector_savings_k_rounded = np.rint(vector_savings_k).astype(int)
```

```python
plt.plot(rounded,label='wartość mieszkania w 5 lat')
plt.plot(vector_savings_k_rounded,label='moje oszczedności w 5 lat')
plt.legend()
plt.xlabel('Liczba okresów')
plt.ylabel('Wartość')
```
